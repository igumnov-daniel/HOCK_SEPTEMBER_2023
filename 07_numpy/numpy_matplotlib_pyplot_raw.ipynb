```python
x = np.linspace(-4, 4, 100)
y = np.sin(x)
y2 = x
y3 = x ** 2 - 2

plt.plot(x, y, ':r', label="$y = \sin{x}$")
plt.plot(x, y2, '-.g', label="$ y=x $")
plt.plot(x, y3,  label='$ y =x^2 - 2 $', alpha=0.33)

plt.title("Наш первый график")
plt.ylabel("Y")
plt.xlabel("X")
plt.legend()

plt.grid()

plt.show()
```

```python
x = np.linspace(-4, 4, 100)
y = np.sin(x)
y2 = x
y3 = x ** 2 - 2

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.title("Наш первый график")
plt.plot(x, y, ':r', label="$y = \sin{x}$")

plt.subplot(1, 3, 2)
plt.title("Наш ВТОРОЙ график")
plt.plot(x, y2, '-.g', label="$ y=x $")

plt.subplot(1, 3, 3)
plt.title("Наш третий график")
plt.plot(x, y3,  label='$ y =x^2 - 2 $', alpha=0.5)

# plt.ylabel("Y")
# plt.xlabel("X")
# plt.legend()

plt.grid()

plt.show()
```

```python
x = np.linspace(-4, 4, 100)
y = np.sin(x)
y2 = x
y3 = x ** 2 - 2
y4 = np.tan(x)

fig, ax = plt.subplots(2, 2, figsize=(10, 8), dpi=100)

ax[0, 0].plot(x, y)
ax[0, 0].set_title('sin')
ax[0, 0].set_xticks([-np.pi / 2, np.pi / 2])
ax[0, 0].set_xticklabels(['$ -\\frac{\\pi}{2}  $', '$ \\frac{\\pi}{2}$ '])

ax[1, 0].plot(x, y2)
ax[1, 0].set_title('linear')

ax[1, 1].bar([1,2,3], [10, 5, 1])
ax[1, 1].set_xticks([1, 2, 3])
ax[1, 1].set_xticklabels(["первый", "второй", "третий"])
ax[1, 1].set_title('quad')

ax[0, 1].plot(x, y4)
ax[0, 1].set_title('tan')

ax[0, 0].tick_params(which='major', width=10, length=10)

plt.show()
```

# Немножко теории вероятностей и статистики

Посчитаем, насколько скоррелированы акции, например, гугла, со стоимостью золота.  
Воспользуем обычной линейной корреляцией Пирсона:  
<img src="https://editor.analyticsvidhya.com/uploads/39170Formula.JPG" />

```python
gold = yf.download('GC=F', start_date, end_date)['Adj Close'].values
gold[-5:] 
```

```python
# построим график стоимости акций гугла и золота
```

```python
x = google[1:] # нужна убрать 1 число, так как цена золота доступа с 3 января, а акций -- со 2 января.
y = gold
```

```python
# посчитаем коэф.кор.
numerator = ((x - x.mean()) * (y - y.mean())).sum()
numerator
```

```python
denominator = np.sqrt((np.power(x - x.mean(), 2).sum() * np.power(y - y.mean(), 2)).sum())
denominator
```

```python
(numerator / denominator).round(3)
```

```python
(numerator / (x.std() * y.std() * x.size)).round(3)
```
