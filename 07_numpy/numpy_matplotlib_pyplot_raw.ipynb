```python
x = np.linspace(-4, 4, 100)
y = np.sin(x)
y2 = x
y3 = x ** 2 - 2

plt.plot(x, y, ':r', label="$y = \sin{x}$")
plt.plot(x, y2, '-.g', label="$ y=x $")
plt.plot(x, y3,  label='$ y =x^2 - 2 $', alpha=0.33)

plt.title("  ")
plt.ylabel("Y")
plt.xlabel("X")
plt.legend()

plt.grid()

plt.show()
```

```python
x = np.linspace(-4, 4, 100)
y = np.sin(x)
y2 = x
y3 = x ** 2 - 2

fig, ax = plt.subplots(2, 2)

ax[0, 0].plot(x, y)
ax[1, 0].plot(x, y2)
ax[1, 1].plot(x, y3)

plt.show()
```

```python
x = np.linspace(-4, 4, 100)
y = np.sin(x)
y2 = x
y3 = x ** 2 - 2

plt.figure(figsize=(10, 10))

plt.subplot(2, 1, 1)
plt.plot(x, y)
plt.title('sin')

plt.subplot(2, 2, 3)
plt.plot(x, y2)
plt.title('line')

plt.subplot(2, 2, 4)
plt.plot(x, y3)
plt.title('quadratic')

plt.show()
```

```python
x = np.linspace(-4, 4, 100)
y = np.sin(x)
y2 = x
y3 = x ** 2 - 2
y4 = np.tan(x)

fig, ax = plt.subplots(2, 2, figsize=(10, 8), dpi=100)

ax[0, 0].plot(x, y)
ax[0, 0].set_title('sin')
ax[0, 0].set_xticks([-np.pi / 2, np.pi / 2])
ax[0, 0].set_xticklabels(['$ -\\frac{\\pi}{2}  $', '$ \\frac{\\pi}{2}$ '])

ax[1, 0].plot(x, y2)
ax[1, 0].set_title('linear')

ax[1, 1].bar([1,2,3], [10, 5, 1])
ax[1, 1].set_xticks([1, 2, 3])
ax[1, 1].set_xticklabels(["", "", ""])
ax[1, 1].set_title('quad')

ax[0, 1].plot(x, y4)
ax[0, 1].set_title('tan')

ax[0, 0].tick_params(which='major', width=10, length=10)

plt.show()
```
