# Условия

## Только `if`

```python
x = int(input())

if x > 0:
    print('POSITIVE')
```

## `if - else`

```python
x = int(input())

if x > 0:
    print('POSITIVE')
else:
    print('NEGATIVE')
```

## `if - elif - else`

```python
x = int(input())

if x > 0:
    print('POSITIVE')
elif x == 0:
    print('EQUAL TO ZERO')
else:
    print('NEGATIVE')
```

## 100500 `elif` подряд 

```python
x = int(input())

if x > 0:
    print('POSITIVE')
elif x == 0:
    print('EQUAL TO ZERO')
elif x < -100:
    print('VERY VERY NEGATIVE')
else:
    print('NEGATIVE')
```

## Проверка нескольких условий

```python
n = int(input())

if n > 5:
    print(1)
elif n > 10:
    print(2)
elif n == 20:
    print(3)
else:
    print(4)
```

## Вложенные условия

```python
n = int(input())

if n > 0:
    res = n % 2
    if res == 0:
        print('EVEN')
    else:
        print('ODD')
else:
    print('NEGATIVE')
```

```python
n = int(input())
if n > 0 and n % 2 == 0:
    print('ЧЕТНОЕ ПОЛОЖИТЕЛЬНО')
elif n > 0 and n % 2 != 0:
    print('НЕЧЕТНОЕ ПОЛОЖИТЕЛЬНОЕ')
else:
    print('ОТРИЦАТЕЛЬНОЕ НЕПОНЯТНО КАКОЕ')
```
