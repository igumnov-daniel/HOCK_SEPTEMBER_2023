# Основы программирования 

## Переменная
Краеугольная сущность в python -- **переменная**.<br>
Важно понимать и помнить, что переменная -- это не сам объект, а всего лишь ссылка на него.<br>
Таким образом, переменную можно переопределить: сначала сказать, что переменная -- это число 10, а потом, что слово "жаба".<br><br>
<img src="./pics/переменная.jpg" width=35%/>
<br>
Значение переменной присваивается через знак равенства. 

```python
var = 10
var = "жаба"
```

Здесь первым и последний раз замечу, что "присвоение значения" -- некорректная формулировка.<br>
На самом в памяти создаётся объект, на который переменная ссылается.<br>
То есть переменная `var` сверху сначала ссылалась на целое число, потому на строку.<br>
Когда на объект в памяти никто не ссылается, он автоматически удаляется из памяти.

## Функция и зарезервированные имена

Для осуществления операций есть соответствующие функции. 
Например, чтобы программа вывела что-то на экран, используется функция `print`,<br>
чтобы принять от пользовтеля строку нужна функция `input`, получить тип данных переменной* -- функция `type`.<br>
_*на самом деле получить тип данных объекта, на который ссылается переменная. но опускаем эти детали, они для нас не так важны_

```python
var = 'жаба'  # создание переменной
print(var)    # печать на экран
              # пропуск строки для красоты. не влияет на программу
t = type(var) # получаем тип переменной и сохраняем в переменную
print(t)      # печатаем тип данных переменной var
```

Наверное, вы уже обратили внимание, что какие-то слова подсвечиваются. Например, названия функций у вас, вероятно, окрашены в зеленый цвет.<br>
Такой зеленый цвет свидетельствует о том, что это имя зарезервировано, то есть о том, что слову `type` соответсвует какая-то функция.<br>
Или, иными словами, что в памяти есть функция, а переменная `type` нас на эту функцию отсылает.<br>
Мы можем передать функции в скобочках аргументы для её работы. И получить результат.<br>
<br>
**НО** если мы переобределим, на что ссылается функция `type`, то потеряем доступ к функции.<br>
Ведь теперь переменная будет ссылаться не на функцию, а на то, что мы сказали.

```python
var = 'жжжжаба'
t = type(var)
print(t)
```

Переобределим значение, на которое ссылается `type`. Теперь это просто число.<br>
Таким образом, теперь `type(var)` это не запуск функции определния типа над переменной `var`,<br>
а это буквально запись `12345(var)`. Мы пытается запустить число 12345. Для числа эта операция не определена.<br><br>
Получаем соответствующую ошибку, что число запустить нельзя (запустить==call, поэтому ошибка  `'int' object is not callable`) 

```python
type = 12345
print(type(var))
```

Чтобы вернуть всё, как было, надо перезагрузить ядро (кнопка restard kernel сврху на панели, стрелочка круговая).

## Типы данных

Первые типы данных, с которыми мы познакомимся:
* **int** (integer), целое число;
* **str** (string), строка;
* **float**, число с плавающей точкой (нецелые собственно);
* **bool**, булеове значение (Истина/Ложь)

### int

#### доступные операции

```python
num = 10
num2 = 4
print(num + num2) # сложение
print(num - num2) # вычитание
print(num * num2) # умножение
print(num / num2) # деление
```

```python
# а также
print(num // num2) # делочисленное деление (div)
print(num % num2)  # определение остатка от деления (mod)
```

```python
# конвертирование других типов данных в целое число 
not_num = '100'
print(not_num)
print(type(not_num)) # type не сработает, если вы не перезапустили ядро
```

```python
num = int(not_num) # конвертируем not_num в целое число
print(num)
print(type(num))
```

```python
# если нельзя сконвертировать, то будет ошибка
not_num = 'HOCK'
num = int(not_num)
```

## Взаимодействие с пользователем
Мы можем получать от пользователя ввод с помощью функции `input`.<br>
Функция всегда возвращает строковый тип данных, что бы ни было введено.

```python
data = input()
print(type(data))
```

```python
data = input()
print(type(data))
```

```python
# если мы знаем, что будет введено число, то надо сконвертирвать значение в числовое
```

```python
data = input()
print(type(data))
print(data * 5) # умножим СТРОКУ на 5 

data = int(data)
print(type(data))
print(data * 5) # умножим ЧИСЛО на 5 
```

```python
# можно также сконвертировать и float
data = input()
print(type(data))
print(data * 5) # умножим СТРОКУ на 5 

data = float(data)
print(type(data))
print(data * 5) # умножим ЧИСЛО (float) на 5 
```

Обратите внимание, `float` в виде строки нелья сконвертировать в целое число,<br>
Нужно сначала сконвертировать во `float`, а потом в `int`.

```python
# так будет ошибка
data = input()
data = int(data)
```

```python
# так ошибка не будет
data = input()
print(data)
print(type(data))  # сейчас это строка 

data = float(data)
print(data)
print(type(data))  # сейчас это float 

data = int(data)
print(data)
print(type(data))  # сейчас это целое число # дробную часть удалили, так как оно целое. 
```
