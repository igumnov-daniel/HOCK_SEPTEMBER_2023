```python
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_diabetes
from statsmodels.regression.linear_model import OLS
```

```python
X, y = load_diabetes(return_X_y=True, as_frame=True, scaled=False)
```

```python
X.head()
```

```python
X.shape
```

```python
y
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
X['constant'] = 1
```

```python
model2 = OLS(y, X)
```

```python
res = model2.fit()
```

```python
print(res.summary())
```

```python
(model.coef_.round(5) != res.params.values[:-1].round(5)).sum()
```

```python
X_scaled = (X - X.mean()) / X .std()
X_scaled['constant'] = 1
```

```python
model2 = OLS(y, X_scaled)
res = model2.fit()
print(res.summary())
```
