```python
x = google[2:] # нужна убрать 1 число, так как цена золота доступа с 3 января, а акций -- со 2 января.
y = gold
```

```python
# посчитаем коэф.кор.
numerator = ((x - x.mean()) * (y - y.mean())).sum()
numerator
```

```python
denominator = np.sqrt((np.power(x - x.mean(), 2).sum() * np.power(y - y.mean(), 2)).sum())
denominator
```

```python
(numerator / denominator).round(3)
```
