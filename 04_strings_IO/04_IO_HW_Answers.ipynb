```python
with open("./data/menu.csv", "r", encoding='utf-8') as menu:
    pair = menu.readline().split(",")
    user_dish = pair[0]
    user_dish_num = int(pair[1])
    
    menu.readline()
    for line in menu:
        line = line.strip().split(';')
        line_dish = line[0]
        line_dish_ccal = int(line[-1])
        
        if line_dish == user_dish:
            res = user_dish_num * line_dish_ccal
            break
            
print(res)
```

```python
with open("./data/grades.csv", 'r', encoding='utf-8') as inpfile:
    with open("output.txt", 'w', encoding='utf8') as outfile:
        header = inpfile.readline().split(',')
        outfile.write('{}\t{}\t{}\t"Mark"\n'.format(
            header[0].strip(),
            header[1].strip(),
            header[2].strip()
        ))
        
        for line in inpfile:
            line = line.strip().split(',')
            nums = line[3:]
            
            for i, elem in enumerate(nums):
                nums[i] = float(nums[i]) # float(elem)
            
            mean = sum(nums) / len(nums)
            
            if mean >= 90:
                mark = 'A'
            elif mean >= 75:
                mark = 'B'
            elif mean >= 60:
                mark = 'C'
            elif mean >= 50:
                mark = 'D'
            else:
                mark = 'F'
                
            outfile.write("{}\t{}\t{}\t{}\n".format(
                line[0].strip(),
                line[1].strip(),
                line[2].strip(),
                mark
            ))
```

```python
with open("./data/weather_stats/sunshine.csv", 'r', encoding='utf-8') as inpf:
    with open("output.txt", 'w', encoding='utf8') as outf:
        inpf.readline()
        
        for line in inpf:
            line = line.strip().split(';')
            city_and_state = line[0]
            nums = line[1:]
            
            #   
            for i in range(len(nums)):
                nums[i] = int(nums[i])
            #  
            s = 0
            q = 0
            for n in nums:
                s += n
                q += 1
            mean = s / q 
            
            # outf.write(city_and_state.ljust(30) + '\t' + str(round(mean, 1)).rjust(6) + '\n')
            # outf.write("{}\t{}\n".format(city_and_state.ljust(30), str(round(mean, 1)).rjust(6)))
            # outf.write(f"{city_and_state:<30}\t{mean:>6.1f}\n")
            outf.write("{:<30}\t{:>6.1f}\n".format(city_and_state, mean))
            
```
