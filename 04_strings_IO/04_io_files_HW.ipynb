Создайте файл `unique_states.txt` с уникальными названиями штатов, отсортированными в лексикографическом порядке по возрастанию (каждый город на новой строке).

Подсказки:
* Проитерируйтесь файл;
* Вычлените из строки с названием города и штата только штат;
* Сохраняйте штат, который вы встретили впервые, в список (то есть нужно как-то проверять, а встречали ли мы штат до этого);
* Отсортировать список поможет функция `sorted` или метод списков `sort` (функция `sorted` возвращает отсортирвоанную копию, а метод `sort` сортирует in place и ничего не возвращает, примеры ниже)
* Пройтись по списку с уникальными штатами и записать каждый в новый файлик.

```python
# пример c sorted
letters = ['B', 'A', 'C']     # создаем исходный спиок
letters_new = sorted(letters) # сортируем
print(letters)                # печатаем исходный список
print(letters_new)            # печатаем новый отсортированный список
```

```python
# пример c sort
letters = ['B', 'A', 'C']     # создаем исходный спиок
letters_new = letters.sort()  # сортируем. Метод НИЧЕГО не возвращает. Переменная letters_new ПУСТАЯ!!!
print(letters)                # печатаем исходный список (он отсортирован)
print(letters_new)            # метод sort вернул НИЧЕГО, поэтому в переменной НИЧЕГО (None)
```
